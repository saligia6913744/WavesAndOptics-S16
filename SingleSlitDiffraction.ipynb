```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Define variables that will be used for the theoretical curve:

```python
theta = np.linspace(-0.03,0.03, num=1000)
beta = 0.5*(2*np.pi/(632.8e-9))*0.08e-3*np.sin(theta)
```

Put data into lists:

```python
ang = [-0.028,-0.025,-0.022,-0.019,-0.017,-0.014,-0.011,-0.008,-0.006,-0.003,0.,
     0.003,0.006,0.008,0.011,0.014,0.017,0.019,0.022,0.025,0.028]
I = [0.0011,0.0075,0.0108,0.0054,0.0172,0.0517,0.0409,0.2788,0.8504,0.8719,1.,
     0.8717,0.8504,0.2788,0.0409,0.0517,0.0172,0.0054,0.0108,0.0075,0.0011]

sigmaAng = [0.001]*25
sigmaI = [0.025]*25
```

Now plot theoretical curve and put data on top:

```python
plt.plot(theta,(np.sin(beta)/beta)**2,"black" ,label="theory")
plt.xlim(-0.03, 0.03)
plt.ylim(-0.1,1.1)

plt.errorbar(ang,I,yerr=sigmaI,xerr=sigmaAng,fmt='none',label="exp. error")
plt.scatter(ang,I,s=10,c="r",marker='o',edgecolors='face',label="data")
plt.legend(frameon=True, loc="best")

plt.ylabel("$I/I_{0}$")
plt.xlabel("Angular Displacement (rad)")
plt.title("Single Slit Diffraction Pattern")
```
