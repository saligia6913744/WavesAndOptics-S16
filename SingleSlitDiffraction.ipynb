```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Define variables that will be used for the theoretical curve:

```python
theta = np.linspace(-0.03,0.03, num=1000)
beta = 0.5*(2*np.pi/(632.8e-9))*0.08e-3*np.sin(theta)
```

Put data into lists:

```python
ang = [-0.024,-0.022,-0.02,-0.018,-0.016,-0.014,-0.012,-0.01,-0.008,-0.006,-0.004,-0.002,0.,
     0.002,0.004,0.006,0.008,0.010,0.012,0.014,0.016,0.018,0.02,0.022,0.024]
I = [0.0862,0.0568,0.0511,0.0398,0.0455,0.0568,0.0909,0.0568,0.0739,0.1648,0.3466,0.5795,
     1.,0.5795,0.3466,0.1648,0.0739,0.0568,0.0909,0.0568,0.0455,0.0398,0.0511,0.0568,0.0682]

sigmaAng = [0.001]*25
sigmaI = [0.025]*25
```

Now plot theoretical curve and put data on top:

```python
plt.plot(theta,(np.sin(beta)/beta)**2,"black" ,label="theory")
plt.xlim(-0.03, 0.03)
plt.ylim(-0.1,1.1)

plt.errorbar(ang,I,yerr=sigmaI,xerr=sigmaAng,fmt='none',label="exp. error")
plt.scatter(ang,I,s=10,c="r",marker='o',edgecolors='face',label="data")
plt.legend(frameon=True, loc="best")

plt.ylabel("$I/I_{0}$")
plt.xlabel("Angular Displacement (rad)")
plt.title("Single Slit Diffraction Pattern")
```
